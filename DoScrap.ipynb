```python
from newspaper import Article
from newspaper import Config
import nltk
nltk.download('punkt')

from transformers import pipeline
import gradio as gr
from gradio.mix import Parallel, Series
```

```python
USER_AGENT = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'

config = Config()
config.browser_user_agent = USER_AGENT
config.request_timeout = 10
```

```python
import bs4
collections=[]
def get_html(url):
    r = requests.get(url)
    return r.text
def get_soup(html):
    soup = bs4.BeautifulSoup(html, 'html.parser')
    return soup

def constants():
  global collections
```

```python
pagesToGet=100
upperframe=[]
linklist=[]
for page in range(0,pagesToGet+1):
  print('Processing Page :', page)
  url='https://netl.doe.gov/news-room-news-stories?tid=All&year=All&page='+str(page)

  try:
    page=requests.get(url)

  except Exception as e:
    error_type,error_obj,error_info= sys.exc_info()
    print('Error For Link:', url)
    print(error_type,'Line:', error_info().tb_lineno)
    continue

  time.sleep(2)

  soup= BeautifulSoup(page.content , 'html.parser')

  frame=[]


  links= soup.find_all('div', attrs={'class':'dcorner-row views-row'})
  print(len(links))


  filename="Webscrapped.csv"
  f=open(filename,"w")
  headers="Statement, Link \n"
  f.write(headers)

  for j in links:
    Statement = j.find('div', attrs={'class':'field-content dcorner-title'}).text
    Link= j.find('div', attrs={'class':'field-content dcorner-title'}).find('a')['href'].strip()
    linklist.append('https://netl.doe.gov'+ Link)
    #Date= j.find('div', attrs={'class':'dcorner-row views-row'}).find('a')['footer'].text[-14:-1].strip()
    #Source= j.find('div', attrs={'class':'dcorner-row views-row'}).find('a').text.strip()

    frame.append((Statement,Link))
    f.write(Statement)

  upperframe.extend(frame)
f.close()

data= pd.DataFrame(upperframe, columns=['Statement','Link'])
data.head(10)


```

```python
def fetch_date(date):
  Date= date.strip('\n')
  Date = datetime.datetime.strptime(Date, '%B %d, %Y').date()
  return Date
```

```python
def save_posts():
    constants()
    with open("collections.json","w") as f:
        json.dump(collections,f,indent=4)
```

```python
def get_sum(each_link):
  url = each_link
  article = Article(url , config=config)
  article.download() 
  article.html
  article.parse() 

  article.nlp()
  return article.summary
```

**Scrapper**

```python
def scrapper():
  constants()
  i=0
  for each_link in linklist:
    html = get_html(each_link)
    post_soup = get_soup(html)
    sum_art= get_sum(each_link)
    
    try:
        title = post_soup.find('div',{'class':'h2 story-title'}).text
        desc = post_soup.find('div',{'class':'story-text'}).text
        extdate = post_soup.find('div',{'class':'post-date2'}).text
        publish_date= fetch_date(extdate)
        
        
        collections.append({'publish_date': str(publish_date),
                    'source': "NETL",
                    'title':title.strip('\n'),
                    'description':desc.strip('\n'),
                    'summary':sum_art.strip('\n')
                    })
        i+=1
        print("Scraped : ", title)
    except Exception as e:
        print("Scraping Error :",e)

```
